```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def lpf(x, pass_freq=10., cut_freq=12., sample_rate=1000):
  wp = pass_freq/(sample_rate/2)
  ws = cut_freq/(sample_rate/2)
  N, Wn = signal.buttord(wp, ws, 1, 20, False)
  b, a = signal.butter(N, Wn, btype='low', analog=False, output='ba')
  y = signal.lfilter(b, a, x)
  return y
```

```python
N = 1000
dt = 0.001
t = np.linspace(1, N, N) * dt - dt
y = np.cos(2*np.pi*t) + np.cos(10*np.pi*t) + np.cos(100*np.pi*t)

plt.plot(t, y)
```

```python
y_ = lpf(y)
plt.plot(t, y_)
```
