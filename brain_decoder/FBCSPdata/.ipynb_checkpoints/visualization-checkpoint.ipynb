```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
Xlr = np.load('task(lr)model(FBCSP)_acc.npy')
Xhf = np.load('task(hf)model(FBCSP)_acc.npy')
Xlh = np.load('task(lh)model(FBCSP)_acc.npy')
Xrh = np.load('task(rh)model(FBCSP)_acc.npy')
Xlf = np.load('task(lf)model(FBCSP)_acc.npy')
Xrf = np.load('task(rf)model(FBCSP)_acc.npy')
Xlra = np.load('atask(lr)model(FBCSP)_acc.npy')
Xhfa = np.load('atask(hf)model(FBCSP)_acc.npy')
Xlha = np.load('atask(lh)model(FBCSP)_acc.npy')
Xrha = np.load('atask(rh)model(FBCSP)_acc.npy')
Xlfa = np.load('atask(lf)model(FBCSP)_acc.npy')
Xrfa = np.load('atask(rf)model(FBCSP)_acc.npy')
```

```python
X = np.vstack((Xlr,Xhf,Xlh,Xrh,Xlf,Xrf))
Xa = np.vstack((Xlra,Xhfa,Xlha,Xrha,Xlfa,Xrfa))
Xa.shape, X.shape
Xnew = np.hstack([X,Xa])
task_mean = Xnew.mean(axis=0)
sub_mean = Xnew.mean(axis=1)
all_mean = Xnew.mean()
print(Xnew.shape)
print(sub_mean.shape)
print(task_mean.shape)
print(all_mean)
mean = np.hstack([task_mean, all_mean])
X_data = np.hstack([Xnew, sub_mean.reshape(-1,1)])
print(X_data.shape)
print(mean.shape)
X_data = np.vstack([X_data, mean.reshape(1,-1)])
print(X_data.shape)
```

```python
df = pd.DataFrame(X_data, columns=['subject'+str(i) for i in range(1,100)]+['subject平均'],
                 index=['LR', 'HF', 'LH', 'RH', 'LF', 'RF','タスク平均']).T
df['subject'] = ['subject'+str(i) for i in range(1,100)]+['subject平均']
df.drop('subject',axis=1)
```
