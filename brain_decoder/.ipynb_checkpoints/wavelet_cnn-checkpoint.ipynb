```python
# class Wavelet_cnn(nn.Module):
#     '''
#     input size is (batch, 1, seq, electrode)
#     output[0] size is (batch, level, seq/2 , electrode)
#     output[1] is list of coefs
#     '''
#     def __init__(self, conv_length=2, stride=2, level=100):
#         super(Wavelet_cnn, self).__init__()
#         if conv_length % 2 == 0:
#             pad = 2*int(conv_length/2)-1
#         else:
#             pad = 2*int(conv_length/2)
#         self.pad = nn.ZeroPad2d((0, 0, pad, 0))
#         self.conv_length = conv_length
#         self.level = level
#         self.l_conv = nn.Conv2d(1, 1, (conv_length, 1), stride=(1, 1), bias=False)
#         self.h_conv = nn.Conv2d(1, 1, (conv_length, 1), stride=(1, 1), bias=False)
#         self.l_downsample = nn.Conv2d(1, 1, (1, 1), stride=(stride, 1), bias=False)
#         self.h_downsample = nn.Conv2d(1, 1, (1, 1), stride=(stride, 1), bias=False)
#         self.recon = nn.ConvTranspose2d(1, 1, (stride, 1), stride=(stride, 1), bias=False)

#     def forward(self, x):
#         b = x
#         a_list = []
#         a_coff = []
#         for i in range(self.level):
#             a = self.h_downsample(self.h_conv(self.pad(b)))
#             b = self.l_downsample(self.l_conv(self.pad(b)))
#             a_coff.append(a)
# #             print(a.size())
#             for j in range(i):
#                 a = self.recon(a)
#             print(a.size())
#             a_list.append(a)

#             if b.size(2) < self.conv_length:
#                 break
#         print('composition level is {}'.format(i+1))
#         return th.stack(a_list, dim=2).squeeze(1), a_coff
class Wavelet_cnn(nn.Module):
    '''
    input size is (batch, 1, seq, electrode)
    output[0] size is (batch, level, seq/2 , electrode)
    output[1] is list of coefs
    '''
    def __init__(self, conv_length=2, stride=2, level=100):
        super(Wavelet_cnn, self).__init__()
        pad = conv_length - 1
        self.pad = nn.ZeroPad2d((0, 0, pad, 0))
        self.conv_length = conv_length
        self.level = level
        ### l_pass, h_pass
        self.l_conv = nn.Conv2d(1, 1, (conv_length, 1),
                                stride=(1, 1), bias=False)
        self.h_conv = nn.Conv2d(1, 1, (conv_length, 1),
                                stride=(1, 1), bias=False)
        ### down sampler
        self.l_downsample = nn.Conv2d(1, 1, (1, 1),
                                      stride=(stride, 1), bias=False)
        self.h_downsample = nn.Conv2d(1, 1, (1, 1),
                                      stride=(stride, 1), bias=False)
        ### up sampler
        self.upsample = nn.ConvTranspose2d(1, 1, (stride, 1),
                                        stride=(stride, 1), bias=False)

    def init_filter(self):
        pass
        
    def forward(self, x):
        b = x
        a_list = []
        a_coff = []
        for i in range(self.level):
            a = self.h_downsample(self.h_conv(self.pad(b)))
            b = self.l_downsample(self.l_conv(self.pad(b)))
            a_coff.append(a)
            print(a.size())
            for j in range(i):
                a = self.upsample(a)
#             print(a.size())
            a_list.append(a)

            if b.size(2) < self.conv_length:
                break
        print('composition level is {}'.format(i+1))
        return th.stack(a_list, dim=2).squeeze(1), a_coff


class Mydnn(nn.Module):
  def __init__(self, time_window=2, stride=2, n_elec=64):
    super(Mydnn, self).__init__()
    self.wavelet = Wavelet_cnn(time_window, stride)
    self.conv1 = nn.Conv2d(n_elec, 128, (25, 2))
    self.conv2 = nn.Conv2d(128, 256, (25, 2))
    self.conv3 = nn.Conv2d(256, 512, (25, 2))
    self.pool = nn.AvgPool2d((2, 1))
    self.conv_class = nn.Conv2d(512, 1024, (1, 1))
    self.pool_class = nn.AdaptiveAvgPool2d(output_size=(1, 1))
    self.linear = nn.Linear(1024, 100)
    self.dropout = nn.Dropout(0.5)
    self.linear_out = nn.Linear(100, 2)

  def forward(self, x):
    scalegram, _ = self.wavelet(x)
    scalegram = scalegram ** 2 # power
    scalegram = scalegram.transpose(1, 3) # (batch, elec, seq, level)
    h = self.pool(F.relu(self.conv1(scalegram)))
    h = self.pool(F.relu(self.conv2(h)))
    h = self.pool(F.relu(self.conv3(h)))
    h = self.pool_class(self.conv_class(h)).squeeze(-1).squeeze(-1)
    h = F.softmax(self.linear_out(self.dropout(self.linear(h))), dim=1)
    return h
```

```python
from skorch.net import NeuralNetClassifier, NeuralNetRegressor

class Wavelet_decompo(nn.Module):
    '''
    input size is (batch, 1, seq, electrode)
    output[0]: decomposition: size is (batch, 1, seq, electrode)
    output[1]: scalegram: size is (batch, level, seq, electrode)
    output[2]: list of pure coef 
    '''
    def __init__(self, conv_length=32, stride=2, level=100):
        super(Wavelet_decompo, self).__init__()
        pad = conv_length - 1
        self.pad = nn.ZeroPad2d((0, 0, pad, 0))
        self.conv_length = conv_length
        self.stride = stride
        self.level = level
        ### l_pass, h_pass
        self.l_conv = nn.Conv2d(1, 1, (conv_length, 1),
                                stride=(1, 1), bias=False)
        self.h_conv = nn.Conv2d(1, 1, (conv_length, 1),
                                stride=(1, 1), bias=False)
        ### down sampler
        self.l_downsample = nn.Conv2d(1, 1, (1, 1),
                                      stride=(stride, 1), bias=False)
        self.h_downsample = nn.Conv2d(1, 1, (1, 1),
                                      stride=(stride, 1), bias=False)
        ### up sampler
        self.upsample = nn.ConvTranspose2d(1, 1, (stride, 1),
                                        stride=(stride, 1), bias=False)
        self.init_filter()
        
        
    def init_filter(self):
        self.l_conv.weight.data.fill_(1.0*(1/self.l_conv.weight.size(2)))
        self.h_conv.weight.data.fill_(1.0*(1/self.h_conv.weight.size(2)))
        for i in range(0, self.conv_length, 2):
            self.h_conv.weight.data[:,:,i,:] *= -1
    
    def _forward(self, x):
#         x = x.unsqueeze(1).unsqueeze(-1)
        b = x
        a_list = []
        a_coef = []
        for i in range(self.level):
            a = self.h_downsample(self.h_conv(self.pad(b)))
            b = self.l_downsample(self.l_conv(self.pad(b)))
            a_coef.append(a)
#             print(a.size())
            for j in range(i+1):
                a = self.upsample(a)
#             print(a.size())
            a_list.append(a)

            if b.size(2) < self.stride:
                break
#         print('composition level is {}'.format(i+1))
        scalegram = th.stack(a_list, dim=2)
        decomposition = scalegram.sum(2)
        
        return decomposition, scalegram, a_coef
    
    def forward(self, x):
        return self._forward(x)[0]


```

```python
class Wavelet_cnn(nn.Module):
    '''
    input size is (batch, 1, seq, electrode)
    seq is expected 2^n
    output[0]: decomposition: size is (batch, 1, seq, electrode)
    output[1]: scalegram: size is (batch, level, seq, electrode)
    output[2]: list of pure coef
    '''
    def __init__(self, conv_length=32, stride=2, level=100):
        super(Wavelet_cnn, self).__init__()
        pad = conv_length - 1
        self.pad = nn.ZeroPad2d((0, 0, pad, 0))
        self.conv_length = conv_length
        self.stride = stride
        self.level = level
        ### l_pass, h_pass
        self.l_conv1 = nn.Conv2d(1, 1, (conv_length, 1),
                                stride=(1, 1), bias=False)
        self.h_conv1 = nn.Conv2d(1, 1, (conv_length, 1),
                                stride=(1, 1), bias=False)
        ### down sampler
        self.l_downsample1 = nn.Conv2d(1, 1, (1, 1),
                                      stride=(stride, 1), bias=False)
        self.h_downsample1 = nn.Conv2d(1, 1, (1, 1),
                                      stride=(stride, 1), bias=False)

        self.l_conv2 = nn.Conv2d(1, 1, (conv_length, 1),
                                stride=(1, 1), bias=False)
        self.h_conv2 = nn.Conv2d(1, 1, (conv_length, 1),
                                stride=(1, 1), bias=False)
        ### down sampler
        self.l_downsample2 = nn.Conv2d(1, 1, (1, 1),
                                      stride=(stride, 1), bias=False)
        self.h_downsample2 = nn.Conv2d(1, 1, (1, 1),
                                      stride=(stride, 1), bias=False)

    def init_filter(self):
        self.l_conv1.weight.data.fill_(1.0*(1/self.l_conv1.weight.size(2))).cuda()
        self.h_conv1.weight.data.fill_(1.0*(1/self.h_conv1.weight.size(2))).cuda()
        self.l_conv2.weight.data.fill_(1.0*(1/self.l_conv2.weight.size(2))).cuda()
        self.h_conv2.weight.data.fill_(1.0*(1/self.h_conv2.weight.size(2))).cuda()        
        for i in range(0, self.conv_length, 2):
            self.h_conv1.weight.data[:,:,i,:] *= -1
            self.h_conv2.weight.data[:,:,i,:] *= -1
            
    def forward(self, x):
#         x = x.unsqueeze(1).unsqueeze(-1)
        b1 = x
        b2 = x
        a1_coef = []
        a2_coef = []
        for i in range(self.level):
            a1 = self.h_downsample1(self.h_conv1(self.pad(b1)))
            b1 = self.l_downsample1(self.l_conv1(self.pad(b1)))
            a2 = self.h_downsample2(self.h_conv2(self.pad(b2)))
            b2 = self.l_downsample2(self.l_conv2(self.pad(b2)))
            a1_coef.append(a1)
            a2_coef.append(a2)

            if bool(b1.size(2) < self.stride) or bool(b2.size < self.stride):
                break
#         print('composition level is {}'.format(i+1))
        return a1_coef, a2_coef
```

```python
t = th.linspace(0, 10, 1024)
theta = 0.3*t
x1 = th.sin(1*2*np.pi*theta*t).unsqueeze(0) + th.cos(0.4*2*np.pi*theta*t).unsqueeze(0)
x2 = th.sin(1.01*2*np.pi*theta*t).unsqueeze(0) + th.cos(0.2*2*np.pi*theta*t).unsqueeze(0)
x3 = th.sin(0.99*2*np.pi*theta*t).unsqueeze(0) + th.cos(0.8*2*np.pi*theta*t).unsqueeze(0)
X = th.cat((x1,x2,x3,x1,x1,x2,x3,x3,x2,x2)).unsqueeze(1).unsqueeze(-1).transpose(0,-1)
X = th.cat((X,X,X,X,X))
X_fake = X + 0.5 * th.randn(5, 1, 1024, 10)
X_fake.size()
```

```python
model = Wavelet_cnn(conv_length=32, stride=2).cuda()
```

```python
Y = model(X_fake.cuda())
```
